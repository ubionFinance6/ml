```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
iris = load_iris()
iris_df = pd.DataFrame(iris.data,columns=iris.feature_names)
print('feature들의 평균 값')
print(iris_df.mean(),'\n')
print('feature들의 분산 값')
print(iris_df.var())
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("sepal length (cm)")

plt.scatter(iris_df.index,iris_df["sepal length (cm)"])
plt.subplot(2,2,2)
plt.title("sepal width (cm)")
plt.scatter(iris_df.index,iris_df["sepal width (cm)"])
plt.subplot(2,2,3)
plt.title("petal length (cm)")
plt.scatter(iris_df.index,iris_df["petal length (cm)"])
plt.subplot(2,2,4)
plt.title("sepal width (cm)")
plt.scatter(iris_df.index,iris_df["petal width (cm)"])
plt.show()
```

### sklearn.preprocessing.StandardScaler

_class_ sklearn.preprocessing.StandardScaler(_*_,  _copy=True_,  _with_mean=True_,  _with_std=True_)[[source]](https://github.com/scikit-learn/scikit-learn/blob/f3f51f9b6/sklearn/preprocessing/_data.py#L635)[](https://scikit-learn.org/1.1/modules/generated/sklearn.preprocessing.StandardScaler.html?highlight=standardscaler#sklearn.preprocessing.StandardScaler "Permalink to this definition")

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler객체 생성
scaler = StandardScaler()
# StandardScaler로 데이터 셋 변환 , fit()과 transform()호출
iris_scaled = scaler.fit_transform(iris_df)

# transform()시 scale 변환된 dataset이 numpy ndarry로 반환되어 이를 DataFrame으로 변환
iris_df_scaled = pd.DataFrame(data=iris_scaled,columns=iris.feature_names)
print('feature들의 평균 값')
print(iris_df_scaled.mean(),'\n')

print('feature들의 분산 값')
print(iris_df_scaled.var())

plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("sepal length (cm)")
plt.scatter(iris_df_scaled.index,iris_df_scaled["sepal length (cm)"])
plt.subplot(2,2,2)
plt.title("sepal width (cm)")
plt.scatter(iris_df_scaled.index,iris_df_scaled["sepal width (cm)"])
plt.subplot(2,2,3)
plt.title("petal length (cm)")
plt.scatter(iris_df_scaled.index,iris_df_scaled["petal length (cm)"])
plt.subplot(2,2,4)
plt.title("sepal width (cm)")
plt.scatter(iris_df_scaled.index,iris_df_scaled["petal width (cm)"])
plt.show()
```
